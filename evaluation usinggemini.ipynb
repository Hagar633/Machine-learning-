```python
!pip install decord

from pickle import FRAME
from google.colab.patches import cv2_imshow
from decord import VideoReader
import cv2
import numpy as np
from matplotlib import pyplot as plt
import google.generativeai as genai
from PIL import Image
from google.colab import files

uploaded = files.upload()

# If you upload "my_video.mp4", the path is just the filename
video_path = "kid1.mp4"

genai.configure(api_key="AIzaSyAYaaU7WCw_UQJ5ES7Bf2a80fAlpbX1jqk")
model = genai.GenerativeModel("gemini-2.5-flash")

results = {"right": 0, "left": 0, "none": 0}
vr = VideoReader(video_path)              # load the video
frame_count = len(vr)                     # total number of frames
print(f"Total frames: {frame_count}")

fps = 7
step = fps * 2                          # sample every 1 second (7 frames apart)

sample_frame = vr[0].asnumpy()            # grab first frame
rgb = cv2.cvtColor(sample_frame, cv2.COLOR_BGR2RGB)

for i in range(0, frame_count, step):
    frame = vr[i].asnumpy()               # get frame as numpy array
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    # ðŸ‘‰ this is where youâ€™d send the frame to Gemini for analysis
    pil_img = Image.fromarray(rgb)

    response = model.generate_content(
        [pil_img, "In this frame, which leg (right or left) is touching the ground? Reply with 'right', 'left', or 'none'."]
    )
    answer = response.text.strip().lower()

    if answer in results:
        results[answer] += 1

print("Final count:", results)
```
