```python


from ultralytics import YOLO

# Load pre-trained YOLOv8 model (you can use YOLOv5 too)
model = YOLO("yolov8n.pt")

# Run inference on an image
results = model("imagee.jpg")

# Show results


# Get bounding boxes, labels, and scores
for r in results:
  r.show()
  for box in r.boxes:
        print("Class:", r.names[int(box.cls)])
        print("Confidence:", box.conf.item())
        print("Coordinates:", box.xyxy.tolist())  # [x1, y1, x2, y2]
```
